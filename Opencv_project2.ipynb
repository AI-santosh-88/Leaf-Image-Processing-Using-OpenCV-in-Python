## import the packages

```python
import cv2
from google.colab.patches import cv2_imshow
```

## Read & show the image

```python
dir(cv2)
```

```python
image = cv2.imread('/content/leaf.jpg')
```

```python
cv2_imshow(image)
```

```python
image.shape
```

```python
image
```

## Convert into gray scale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2_imshow(rgb_image)
```

```python
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2_imshow(hsv_image)
```

```python
resized_image = cv2.resize(image, (200, 200))
cv2_imshow(resized_image)
```

## Cropping the image

```python
cropped = image[100:300,100:300]
cv2_imshow(cropped)
```

## flip the image

```python
flip_v=cv2.flip(image,0)
flip_h=cv2.flip(image,1)

cv2_imshow(flip_v)
cv2_imshow(flip_h)
```

```python
flip_v=cv2.flip(image,0)
flip_h=cv2.flip(image,1)
flip_hv=cv2.flip(image,-1)
cv2_imshow(flip_v)
cv2_imshow(flip_h)
cv2_imshow(flip_hv)
```
